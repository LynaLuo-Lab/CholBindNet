```python
import pandas as pd
import os

# Load CSV files
cnn3d_pos_df = pd.read_csv("3DCNNExp1Results/test_positive_labels.csv")
gnn_pos_df = pd.read_csv("GNNExp1Results/test_positive_labels.csv")
gat_pos_df = pd.read_csv("GatModelExp1Results/test_positive_labels.csv")

# Function to extract protein ID from filename
def extract_protein_id(path):
    base = os.path.basename(path)
    return base.split("-")[0]  # e.g., "1LRI" from "1LRI-filtered_graphs.npy"

# Add protein_id column
for df in [cnn3d_pos_df, gnn_pos_df, gat_pos_df]:
    df['protein_id'] = df['filename'].apply(extract_protein_id)

# Filter for average_score >= 0.9
cnn3d_filtered = cnn3d_pos_df[cnn3d_pos_df['average_score'] >= 0.9]
gnn_filtered = gnn_pos_df[gnn_pos_df['average_score'] >= 0.9]
gat_filtered = gat_pos_df[gat_pos_df['average_score'] >= 0.9]

# Get sets of protein_ids
cnn_ids = set(cnn3d_filtered['protein_id'])
gnn_ids = set(gnn_filtered['protein_id'])
gat_ids = set(gat_filtered['protein_id'])

# Find overlapping protein IDs
overlapping_ids = cnn_ids & gnn_ids & gat_ids

print(len(overlapping_ids), "is number of overlapping id's")

# Print results
print("Overlapping protein IDs with average_score >= 0.9 in all models:\n")
for protein_id in sorted(overlapping_ids):
    cnn_score = cnn3d_filtered[cnn3d_filtered['protein_id'] == protein_id]['average_score'].values
    gnn_score = gnn_filtered[gnn_filtered['protein_id'] == protein_id]['average_score'].values[0]
    gat_score = gat_filtered[gat_filtered['protein_id'] == protein_id]['average_score'].values[0]
    
    print(f"{protein_id}: 3DCNN = {cnn_score}, GNN = {gnn_score:.4f}, GAT = {gat_score:.4f}")
```
