```python
import pandas as pd

# Load the CSV
df = pd.read_csv("spy_capture_rates.csv")

feature = 'average_score'

# Calculate statistics
min_score = df[feature].min()
max_score = df[feature].max()
median_score = df[feature].median()
mean_score = df[feature].mean()

# Print the results
print(f"Smallest average_score: {min_score:.6f}")
print(f"Largest average_score: {max_score:.6f}")
print(f"Median average_score: {median_score:.6f}")
print(f"Mean average_score: {mean_score:.6f}")

filtered_df = df[(df["number_atoms"] >= 40) & (df["number_atoms"] <= 80)]

# Compute mean of average_score in filtered rows
mean_filtered_score = filtered_df["average_score"].mean()

# Print result
print(f"Mean average_score (number_atoms between 40 and 80): {mean_filtered_score:.6f}")

filtered_df = df[(df["number_atoms"] >= 55) & (df["number_atoms"] <= 65)]

# Compute mean of average_score in filtered rows
mean_filtered_score = filtered_df["average_score"].mean()

# Print result
print(f"Mean average_score (number_atoms between 55 and 65): {mean_filtered_score:.6f}")
```
