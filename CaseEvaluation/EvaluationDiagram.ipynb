```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

gnn_exp1_spies = "../../../Models/Cholesterol/GCN/GCN-5A_Exp1/Results/SpyCaptureRates.csv"

# === Load & prepare ===
df = pd.read_csv(gnn_exp1_spies)
values = df["average_score"].dropna().to_numpy()

median = np.median(values)
p25   = np.percentile(values, 25)   # Q1
p75   = np.percentile(values, 75)   # Q3

# === Plot ===
plt.rcParams.update({"font.size": 18})
fig, ax = plt.subplots(figsize=(7, 7))

# --- Background shading for the 4 regions ---
ax.axhspan(0,     p25,    facecolor="#FF6B6B", alpha=1.0)  # light red-ish
ax.axhspan(p25,   median, facecolor="#FFA36B", alpha=1.0)  # light orange-ish
ax.axhspan(median,p75,    facecolor="#6BFF8D", alpha=1.0)  # light green-ish
ax.axhspan(p75,   1,      facecolor="#6BA3FF", alpha=1.0)  # light blue-ish

# Histogram bars
ax.hist(values, bins="auto", orientation="horizontal",
        edgecolor="black", color="#a6cee3")

# Percentile lines
ax.axhline(p25,    color="#e41a1c", linestyle="--", linewidth=3, label="25th percentile of spies")
ax.axhline(median, color="darkorange", linestyle="-",  linewidth=3, label="Median of spies")
ax.axhline(p75,    color="seagreen",   linestyle="--", linewidth=3, label="75th percentile of spies")

# Labels & styling
ax.set_xlabel("Count")
ax.set_ylabel("Average score")
#ax.set_title("Percentile Labels Sample Based on Spies")
ax.tick_params(axis="both", which="major", labelsize=16, length=7, width=1)

# Legend with white background box
legend = ax.legend(loc="best", frameon=True, fontsize=25)
legend.get_frame().set_facecolor("white")
legend.get_frame().set_edgecolor("black")
legend.get_frame().set_alpha(1.0)

fig.tight_layout()
plt.show()

print(f"Median: {median:.4f} | 25th: {p25:.4f} | 75th: {p75:.4f}")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

gnn_exp1_spies = "../../../Models/Cholesterol/GCN/GCN-5A_Exp1/Results/SpyCaptureRates.csv"

# === Load & prepare ===
df = pd.read_csv(gnn_exp1_spies)
values = df["average_score"].dropna().to_numpy()

mean = np.mean(values)
min   = np.min(values)   # Q1
max   = np.max(values)   # Q3

# === Plot ===
plt.rcParams.update({"font.size": 18})
fig, ax = plt.subplots(figsize=(5, 7))

# --- Background shading for the 4 regions ---
ax.axhspan(0,     min,    facecolor="#cc0000", alpha=0.4)  # light red-ish
ax.axhspan(min,   mean, facecolor="#ff8000", alpha=0.4)  # light orange-ish
ax.axhspan(mean,max,    facecolor="#cccc00", alpha=0.4)  # light green-ish
ax.axhspan(max,   1,      facecolor="#00994d", alpha=0.4)  # light blue-ish

# Histogram bars
ax.hist(values, bins="auto", orientation="horizontal",
        edgecolor="black", color="#a6cee3")

# Percentile lines
ax.axhline(min,    color="#e41a1c", linestyle="--", linewidth=3, label="Min")
ax.axhline(mean, color="darkorange", linestyle="-",  linewidth=3, label="Mean")
ax.axhline(max,    color="seagreen",   linestyle="--", linewidth=3, label="Max")

# Labels & styling
ax.set_xlabel("Count")
ax.set_ylabel("Average score")
#ax.set_title("Min-Max Labels Sample Based on Spies")
ax.tick_params(axis="both", which="major", labelsize=16, length=7, width=1)

# Legend with white background box
legend = ax.legend(loc="best", frameon=True)
legend.get_frame().set_facecolor("white")
legend.get_frame().set_edgecolor("black")
legend.get_frame().set_alpha(1.0)

fig.tight_layout()
plt.show()

print(f"mean: {mean:.4f} | 25th: {min:.4f} | 75th: {max:.4f}")
```
