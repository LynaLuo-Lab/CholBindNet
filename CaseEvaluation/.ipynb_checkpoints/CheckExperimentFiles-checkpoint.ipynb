```python
import glob
import os

exp1_unl_cnn = glob.glob("../../../Data/SplitData/Cholesterol/cholesterol-grid-5A_exp1/Test/Unlabeled/*.npy")
exp1_unl_gnn = glob.glob("../../../Data/SplitData/Cholesterol/cholesterol-graph-5A/Test/Unlabeled/*.npy")

cnn_protein_ids = set()

for file in exp1_unl_cnn:
    cnn_protein_ids.add(os.path.basename(file)[:7])

gnn_protein_ids = set()

for file in exp1_unl_gnn:
    gnn_protein_ids.add(os.path.basename(file)[:7])

print(len(cnn_protein_ids), "is length of cnn ids")
print(len(gnn_protein_ids), "is length of gnn ids")

# IDs in gnn but not in cnn
gnn_not_in_cnn = gnn_protein_ids - cnn_protein_ids
print(f"IDs in GNN but not in CNN: {gnn_not_in_cnn}")

# Overlapping IDs
overlapping_ids = gnn_protein_ids & cnn_protein_ids
print(f"Number of overlapping IDs: {len(overlapping_ids)}")
```
