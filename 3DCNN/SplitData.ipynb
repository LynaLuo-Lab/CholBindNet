```python
import glob
import os
import shutil
import random

# Define the source and destination folders
source_folder = '../../../Data/SplitData/Cholesterol/cholesterol-grid-5A/PositiveWithoutSpies'  # Replace with the path to your source folder
destination_folder = '../../../Data/SplitData/Cholesterol/cholesterol-grid-5A/Spies'  # Replace with the path to your destination folder

# Make sure the destination folder exists
os.makedirs(destination_folder, exist_ok=True)

# Collect all ligand prefixes (first 4 letters of file names)
ligands = {}
for file_name in os.listdir(source_folder):
    if file_name.endswith('.npy'):
        ligand_prefix = file_name[:4]
        ligands.setdefault(ligand_prefix, []).append(file_name)

# Randomly select 154 ligands
selected_ligands = random.sample(list(ligands.keys()), 154)

# Move the selected fragment files to the destination folder
for positive in selected_ligands:
    for file_name in ligands[positive]: 
        src_path = os.path.join(source_folder, file_name)
        dest_path = os.path.join(destination_folder, file_name)
        shutil.move(src_path, dest_path)
        print(f'Moved: {file_name}')

print(f'Moved {len(selected_ligands)} positives to {destination_folder}')
```

```python
source_folder = '../../../Data/SplitData/Cholesterol/cholesterol-grid-5A/UnlabeledWithSpies' 
k_subsets_folder = '../../../Data/SplitData/Cholesterol/cholesterol-grid-5A/k_subsets' 

# Make sure the k_subsets folder exists
os.makedirs(k_subsets_folder, exist_ok=True)

# Get the list of all files in the source folder
all_files = os.listdir(source_folder)

# Create 50 directories (k_subsets)
k = 50
for i in range(k):
    subset_folder = os.path.join(k_subsets_folder, f'subset_{i+1}')
    os.makedirs(subset_folder, exist_ok=True)

# Distribute 385 files to each subset 
for i in range(k):
    # Randomly select 385 files for this subset 
    selected_files = random.sample(all_files, k=(385 * 5))
    
    # Create the subset folder
    subset_folder = os.path.join(k_subsets_folder, f'subset_{i+1}')
    
    # Copy the selected files into the corresponding subset folder
    for file in selected_files:
        src_path = os.path.join(source_folder, file)
        dest_path = os.path.join(subset_folder, file)
        
        # Copy the file
        shutil.copy(src_path, dest_path)
        print(f'Copied {file} to {subset_folder}')

print(f'Finished copying 385 files into each of the {k} subsets.')
```
