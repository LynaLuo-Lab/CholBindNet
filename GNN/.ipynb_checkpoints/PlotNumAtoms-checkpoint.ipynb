```python
import glob
import os
from Bio.PDB import PDBParser
import matplotlib.pyplot as plt

positive_files = glob.glob("filtered-pdbs-distinct-5A/positive/*.pdb")

# Initialize PDB parser
parser = PDBParser(QUIET=True)

# Lists to store counts
total_atom_counts = []
carbon_counts = []
nitrogen_counts = []
oxygen_counts = []

# Process each PDB file
for file in positive_files:
    structure = parser.get_structure("structure", file)
    atoms = list(structure.get_atoms())
    
    total_atom_counts.append(len(atoms))
    carbon_counts.append(sum(1 for atom in atoms if atom.element == 'C'))
    nitrogen_counts.append(sum(1 for atom in atoms if atom.element == 'N'))
    oxygen_counts.append(sum(1 for atom in atoms if atom.element == 'O'))

# Calculate global min and max across all atom counts
all_counts = total_atom_counts + carbon_counts + nitrogen_counts + oxygen_counts
x_min = min(all_counts)
x_max = max(all_counts)

fig, axs = plt.subplots(4, 1, figsize=(10, 12))

# Plot total atom counts
axs[0].hist(total_atom_counts, bins=30, range=(x_min, x_max), edgecolor='black', color='skyblue')
axs[0].set_xlabel("Total Number of Atoms")
axs[0].set_ylabel("Number of PDB Files")
axs[0].set_title("Distribution of Total Atom Counts From Positive Files")
axs[0].grid(True)

axs[1].hist(carbon_counts, bins=30, range=(x_min, x_max), edgecolor='black', color='orange')
axs[1].set_xlabel("Total Number of Carbon")
axs[1].set_ylabel("Number of PDB Files")
axs[1].set_title("Distribution of Carbon Atom Counts From Positive Files")
axs[1].grid(True)

axs[2].hist(nitrogen_counts, bins=30, range=(x_min, 30), edgecolor='black', color='red')
axs[2].set_xlabel("Total Number of Nitrogen")
axs[2].set_ylabel("Number of PDB Files")
axs[2].set_title("Distribution of Nitrogen Atom Counts From Positive Files")
axs[2].grid(True)

axs[3].hist(oxygen_counts, bins=30, range=(x_min, 30), edgecolor='black', color='blue')
axs[3].set_xlabel("Total Number of Oxygen")
axs[3].set_ylabel("Number of PDB Files")
axs[3].set_title("Distribution of Oxygen Atom Counts From Positive Files")
axs[3].grid(True)

plt.tight_layout()
plt.show()
```

```python
import glob
import os
import csv
from Bio.PDB import PDBParser

# Find all positive and unlabeled PDB files
positive_files = glob.glob("filtered-pdbs-distinct-5A/positive/*.pdb")
unlabeled_files = glob.glob("filtered-pdbs-distinct-5A/unlabeled/*.pdb")
all_files = positive_files + unlabeled_files

# Initialize PDB parser
parser = PDBParser(QUIET=True)

# Store basenames of small files
small_files = []

for file in all_files:
    structure = parser.get_structure("structure", file)
    atoms = list(structure.get_atoms())

    if len(atoms) < 20:
        basename = os.path.splitext(os.path.basename(file))[0]
        small_files.append([basename])  # Save as a list to write to CSV
        print(f"Found small file: {basename} ({len(atoms)} atoms)")

# Write small file basenames to a CSV
with open('small_files.csv', 'w', newline='') as csvfile:
    writer = csv.writer(csvfile)
    writer.writerow(['basename'])  # optional header
    writer.writerows(small_files)

print(f"Finished. {len(small_files)} files written to small_files.csv.")
```

```python
unlabeled_files = glob.glob("filtered-pdbs-distinct-5A/unlabeled/*.pdb")
positive_files = glob.glob("filtered-pdbs-distinct-5A/positive/*.pdb")

# Initialize PDB parser
parser = PDBParser(QUIET=True)

# Lists to store counts
positive_total_atom_counts = []
likely_positive_total_atom_counts = []
likely_negative_total_atom_counts = []

for file in positive_files:
    structure = parser.get_structure("structure", file)
    atoms = list(structure.get_atoms())

    positive_total_atom_counts.append(len(atoms))

for file in unlabeled_files:
    structure = parser.get_structure("structure", file)
    atoms = list(structure.get_atoms())
    
    if 'positive' in file.lower():  # case-insensitive check
        likely_positive_total_atom_counts.append(len(atoms))
    else:
        likely_negative_total_atom_counts.append(len(atoms))

# Calculate global min and max across all atom counts
all_counts = positive_total_atom_counts + likely_positive_total_atom_counts + likely_negative_total_atom_counts
x_min = min(all_counts)
x_max = max(all_counts)

fig, axs = plt.subplots(3, 1, figsize=(10, 8))

# Plot total atom counts
axs[0].hist(positive_total_atom_counts, bins=30, range=(x_min, x_max), edgecolor='black', color='skyblue')
axs[0].set_xlabel("Total Number of Atoms")
axs[0].set_ylabel("Number of PDB Files")
axs[0].set_title("Distribution of Total Atom Counts From Positive Files")
axs[0].grid(True)

axs[1].hist(likely_positive_total_atom_counts, bins=30, range=(x_min, x_max), edgecolor='black', color='orange')
axs[1].set_xlabel("Total Number of Atoms")
axs[1].set_ylabel("Number of PDB Files")
axs[1].set_title("Distribution of Total Atom Counts From Likely Positive Files")
axs[1].grid(True)

axs[2].hist(likely_negative_total_atom_counts, bins=30, range=(x_min, x_max), edgecolor='black', color='red')
axs[2].set_xlabel("Total Number of Atoms")
axs[2].set_ylabel("Number of PDB Files")
axs[2].set_title("Distribution of Total Atom Counts From Likely Negative Files")
axs[2].grid(True)

plt.tight_layout()
plt.show()
```

```python
from scipy.stats import gaussian_kde
import numpy as np

positive_files = glob.glob("filtered-pdbs-distinct-5A/positive/*.pdb")
unlabeled_files = glob.glob("filtered-pdbs-distinct-5A/unlabeled/*.pdb")

# Initialize PDB parser
parser = PDBParser(QUIET=True)

# Lists to store counts
positive_total_atom_counts = []
unlabeled_total_atom_counts = []

# Process each PDB file
for file in positive_files:
    structure = parser.get_structure("structure", file)
    atoms = list(structure.get_atoms())
    
    positive_total_atom_counts.append(len(atoms))
    if len(atoms) > 100:
        print(file)

count_over_100 = sum(1 for x in positive_total_atom_counts if x > 100)
print("There are", count_over_100, "elements over 100")

for file in unlabeled_files:
    if "positive" not in file:
        structure = parser.get_structure("structure", file)
        atoms = list(structure.get_atoms())

        unlabeled_total_atom_counts.append(len(atoms))

# Compute KDEs
combined_min = min(min(positive_total_atom_counts), min(unlabeled_total_atom_counts))
combined_max = max(max(positive_total_atom_counts), max(unlabeled_total_atom_counts))
x_vals = np.linspace(combined_min, combined_max, 1000)

positive_kde = gaussian_kde(positive_total_atom_counts)
unlabeled_kde = gaussian_kde(unlabeled_total_atom_counts)

positive_density = positive_kde(x_vals)
unlabeled_density = unlabeled_kde(x_vals)

# Plot
plt.figure(figsize=(10, 6))
plt.plot(x_vals, positive_density, label="Positive Files", color='blue', linewidth=2)
plt.plot(x_vals, unlabeled_density, label="Unlabeled Files", color='orange', linewidth=2)

# Fill overlap
overlap = np.minimum(positive_density, unlabeled_density)
plt.fill_between(x_vals, overlap, color='purple', alpha=0.3, label='Overlap')

# Labels and legend
plt.title("Density of Total Atom Counts in Positive vs. Unlabeled (Likely Negatives) PDB Files")
plt.xlabel("Total Number of Atoms")
plt.ylabel("Density")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
