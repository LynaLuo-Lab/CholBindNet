```python
import pandas as pd

# Load the CSV
df = pd.read_csv("spy_capture_rates.csv")

feature = 'average_score'

# Calculate statistics
min_score = df[feature].min()
max_score = df[feature].max()
median_score = df[feature].median()
mean_score = df[feature].mean()

# Print the results
print(f"Smallest average_score: {min_score:.6f}")
print(f"Largest average_score: {max_score:.6f}")
print(f"Median average_score: {median_score:.6f}")
print(f"Mean average_score: {mean_score:.6f}")
```

```python
# Step 3: Label based on comparison
def assign_label(score):
    if score > max_score:
        return "TruePositive"
    elif score < min_score:
        return "TrueNegative"
    elif score > mean_score:
        return "LikelyPositive"
    else:
        return "LikelyNegative"
```

```python
# Load the CSV
df = pd.read_csv("ivan_capture_rates.csv")

df['label'] = df['average_score'].apply(assign_label)

# Step 4: Save to new CSV
df.to_csv("ivan_labels.csv", index=False)
print("Labeled results saved to labeled_scores.csv")
```

```python
# Load the CSV
df = pd.read_csv("test_unlabeled_capture_rates.csv")

df['label'] = df['average_score'].apply(assign_label)

# Step 4: Save to new CSV
df.to_csv("test_unlabeled_labels.csv", index=False)
print("Labeled results saved to labeled_scores.csv")
```
