```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

pd.set_option('display.width', 200)
pd.set_option('display.max_columns', None)

exp1_pos_mean = [
            [6, 16, 8], 
            [6, 97, 146],
            [12, 0, 0],
            [130, 41, 0]
            ]
exp2_pos_mean = [
            [8, 13, 7], 
            [13, 69, 147],
            [12, 3, 0],
            [121, 69, 0]
            ]
exp3_pos_mean = [
            [6, 29, 16], 
            [12, 78, 138],
            [27, 1, 0],
            [109, 46, 0]
            ]
exp4_pos_mean = [
            [7, 20, 10], 
            [7, 77, 144],
            [17, 0, 0],
            [123, 57, 0]
            ]
exp5_pos_mean = [
            [7, 20, 8], 
            [9, 84, 146],
            [19, 0, 0],
            [119, 50, 0]
            ]
experiments = np.array([
    exp1_pos_mean,
    exp2_pos_mean,
    exp3_pos_mean,
    exp4_pos_mean,
    exp5_pos_mean
])

# Compute mean across the first axis (i.e., average over experiments)
mean_results = np.mean(experiments, axis=0)
std_results = np.std(experiments, axis=0)

print(std_results)

# Optionally, round or convert to int if needed
#mean_results = mean_results.astype(int)  # or .astype(int) if you want integers

# Print the result
labels = ["PseudoNegative", "PseudoPositive", "StrongNegative", "StrongPositive"]
columns = ["Percentile (Using 50th Per)", "Min_Max (Using Spy Cap)", "0.5 Threshold"]

df_mean = pd.DataFrame(mean_results, index=labels, columns=columns)
df_std = pd.DataFrame(std_results, index=labels, columns=columns)
print(df_mean)
print("\n", df_std)

# Plot
fig, ax = plt.subplots(figsize=(8, 6))
x = np.arange(len(labels))
width = 0.25  # Width of each bar

# Plot bars for each column with offset
for i, col in enumerate(columns):
    ax.bar(x + i * width, df_mean[col], width, yerr=df_std[col], capsize=5, label=col, ecolor='red')

# Add labels and title
ax.set_xlabel("Labels")
ax.set_ylabel("Number of Samples")
ax.set_title("3 Case Study Positive Evaluation Set (154 Samples)")
ax.set_ylim(0, 154)
ax.set_xticks(x + width)
ax.set_xticklabels(labels)
ax.legend()

# Show plot
plt.show()
```
